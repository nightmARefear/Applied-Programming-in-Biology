```python
def read_data(file_name):
    f = open(file_name)
    sequence_list = list()
    for line in f:
        seq = line.strip()
        sequence_list.append(seq)
    f.close()
    return sequence_list

def is_nucleotide(x):
    used_nucleotides = ['A', 'T', 'C', 'G']
    x = x.upper()
    if x in used_nucleotides:
        return True
    else:
        False
    
def is_nucleotide_base(baseA, baseB):
    if ((baseA == 'a' or baseA == 'A') and (baseB == 't' or baseB == 'T') or (baseA == 't' or baseA == 'T') and (baseB == 'a' or baseB == 'A')):
        return True
    elif ((baseA == 'c' or baseA == 'C') and (baseB == 'g' or baseB == 'G') or (baseA == 'g' or baseA == 'G') and (baseB == 'c' or baseB == 'C')):
        return True
    else:
        return False
    
def sequence_similarity(seq1, seq2):
    matches = 0
    for i in range(len(seq1)):
        if (seq1[i] == seq2[i]):
            matches += 1
    similarity = matches/len(seq1)
    return similarity
                        
def alignment_similarity(seq1, seq2):
    nucleotide_matches = 0
    single_gaps = 0
    double_gaps = 0
    total_gaps = 0

    for i in range(len(seq1)):
        if (seq1[i] == '-' and  seq2[i] == '-'):
            double_gaps += 1
        elif (seq1[i] == seq2[i]):
            nucleotide_matches += 1
        elif (seq1[i] == '-' or seq2[i] == '-'):
            single_gaps += 1
            
    total_bases = nucleotide_matches + single_gaps
    similarity = nucleotide_matches/total_bases
    return similarity

def get_similarities(unknown, typed_sequences):
        similarities = []
        
        for i in range(len(typed_sequences)):
            similarities.append(alignment_similarity(unknown, typed_sequences[i])*100)
        return similarities

def max_similarity(similarities):
    if type(similarities) == float:
        return similarities
    
    max2 = similarities[0]
    
    for i in range(len(similarities)):
        if similarities[i] > max2:
            max2 = similarities[i]
    return max2

def get_all_similarities(unknown_subtype, known_subtypes):
    
        all_max_similarities = {'A': 0, 'B': 0, 'C': 0, 'D':0} # the max similarities from each sequence of all the subtypes   
        
        for subtype in known_subtypes:
            for sequence in subtype:  
                print('known: ', len(sequence))
                print('\nunknown: ', len(unknown_subtype))
                max_similarities = []  # the similarities of all the sequences of ONE -known- in aspect to ONE unknown subtype sequence
                max_similarity_subtype = 0 # the max similarity of ONE sequence per subtype
            
                max_similarities = get_similarities(unknown_subtype, sequence)
            
            max_similarity_subtype = max_similarity(max_similarities)

            all_max_similarities['A'] = max_similarity_subtype
            if subtype == 0:
                max_similarities['A'] = max_similarity_subtype
            elif subtype == 1:
                max_similarities['B'] = max_similarity_subtype
            elif subtype == 2:
                max_similarities['C'] = max_similarity_subtype
            elif subtype == 3:
                max_similarities['D'] = max_similarity_subtype
                
        return all_max_similarities
            
        
subtypeA_list = read_data('subtypeA.txt')
subtypeB_list = read_data('subtypeB.txt')
subtypeC_list = read_data('subtypeC.txt')
subtypeD_list = read_data('subtypeD.txt')
unknown_list = read_data('unknownType.txt')

all_known_subtypes_list = [subtypeA_list, subtypeB_list, subtypeC_list, subtypeD_list]

similarity = get_similarities(unknown_list[0], subtypeA_list)
print(similarity)

print('\n{}'.format(max_similarity(similarity)))

all_similarities = {}

all_similarities = get_all_similarities(unknown_list[0], all_known_subtypes_list)
print(all_similarities)
```
