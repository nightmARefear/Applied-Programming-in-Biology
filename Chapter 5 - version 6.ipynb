```python
def read_data(file_name):
    f = open(file_name)
    sequence_list = list()
    for line in f:
        seq = line.strip()
        sequence_list.append(seq)
    f.close()
    return sequence_list

def is_nucleotide(x):
    used_nucleotides = ['A', 'T', 'C', 'G']
    x = x.upper()
    if x in used_nucleotides:
        return True
    else:
        False
    
def is_nucleotide_base(baseA, baseB):
    if ((baseA == 'a' or baseA == 'A') and (baseB == 't' or baseB == 'T') or (baseA == 't' or baseA == 'T') and (baseB == 'a' or baseB == 'A')):
        return True
    elif ((baseA == 'c' or baseA == 'C') and (baseB == 'g' or baseB == 'G') or (baseA == 'g' or baseA == 'G') and (baseB == 'c' or baseB == 'C')):
        return True
    else:
        return False
    
def sequence_similarity(seq1, seq2):
    matches = 0
    for i in range(len(seq1)):
        if (seq1[i] == seq2[i]):
            matches += 1
    similarity = matches/len(seq1)
    return similarity
                        
def alignment_similarity(seq1, seq2):
    nucleotide_matches = 0
    single_gaps = 0
    double_gaps = 0
    total_gaps = 0

    '''print(len(seq1))
    print(len(seq2))
    print('\n')'''

    for i in range(len(seq1)):
        if (seq1[i] == '-' and  seq2[i] == '-'):
            double_gaps += 1
        elif (seq1[i] == seq2[i]):
            nucleotide_matches += 1
        elif (seq1[i] == '-' or seq2[i] == '-'):
            single_gaps += 1
            
    total_bases = nucleotide_matches + single_gaps
    similarity = nucleotide_matches/total_bases
    return similarity

def get_similarities(unknown, typed_sequences):
        similarities = []
        
        for i in range(len(typed_sequences)):
            similarities.append(alignment_similarity(unknown, typed_sequences[i])*100)
        return similarities

def max_similarity(similarities):
    if type(similarities) == float:
        return similarities
    
    max2 = similarities[0]
    
    for i in range(len(similarities)):
        if similarities[i] > max2:
            max2 = similarities[i]
    return max2

def get_all_similarities(unknown_subtype, known_subtypes):
    
    all_max_similarities = {'A': 0, 'B': 0, 'C': 0, 'D':0} # the max similarities from each sequence of all the subtypes   
    index = -1 # used instead of range(len(known_subtypes)) since subtype is of type list
    
    for subtype in known_subtypes:
        index += 1
        subtype_similarity_list = []
        subtype_max_similarity = 0
        curr_subtype = 0
        
        for sequence in subtype:
            curr_sequence_similarity = alignment_similarity(unknown_subtype, sequence)
            subtype_similarity_list.append(curr_sequence_similarity)
        
        subtype_max_similarity = max_similarity(subtype_similarity_list)
        
        if index == 0:
            all_max_similarities['A'] = subtype_max_similarity
        elif index == 1:
            all_max_similarities['B'] = subtype_max_similarity
        elif index == 2:
            all_max_similarities['C'] = subtype_max_similarity
        else :
            all_max_similarities['D'] = subtype_max_similarity
            
    return all_max_similarities

def predict_subtype(unknown, all_known_subtypes_list):
    subtypes_similarities = []
    
    for i in range(len(all_known_subtypes_list)):
        print(len(all_known_subtypes_list[i]))
            
        
subtypeA_list = read_data('subtypeA.txt')
subtypeB_list = read_data('subtypeB.txt')
subtypeC_list = read_data('subtypeC.txt')
subtypeD_list = read_data('subtypeD.txt')
unknown_list = read_data('unknownType.txt')

all_known_subtypes_list = [subtypeA_list, subtypeB_list, subtypeC_list, subtypeD_list]

similarity1 = get_similarities(unknown_list[0], subtypeA_list)
similarity2 = get_similarities(unknown_list[0], subtypeB_list)
'''print(similarity1)
print(similarity2)

print('\n{}'.format(max_similarity(similarity1)))
print('\n{}'.format(max_similarity(similarity2)))'''


all_similarities = get_all_similarities(unknown_list[0], all_known_subtypes_list)
print('all similarities: ', all_similarities)

#predict_subtype(unknown_list[0], all_known_subtypes_list)
```
